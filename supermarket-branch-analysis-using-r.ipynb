**VISUALISNG THE DATA **

```r
store%>%
  ggplot(aes(Store_Sales))+
  geom_bar(fill='lightblue')+
  theme_bw()+
  labs(title='group and their Store Sales',
       x = 'Store Sales',
       y = 'group')
```

Group 0 and Group 1 have the lowest sales. Group 2 and Group 3 have the highest sales.****

```r
store%>%
  ggplot(aes(Daily_Customer_Count))+
  geom_bar(fill='lightgreen')+
  theme_bw()+
  labs(title='group and thier Daily Customer Count',
       x = 'Daily Customer Count',
       y = 'group')
```

```r
store%>%
  ggplot(aes(Items_Available, Store_Area ))+
  geom_point(fill='brown',position='jitter')+
  theme_bw()+
  labs(title='items available vs store area',
       x = 'items available',
       y = 'store area')
```

```r
store%>%
  ggplot(aes(Store_Sales, Store_Area ))+
  geom_point(fill='brown',position='jitter')+
  theme_bw()+
  labs(title='store sales vs store area',
       x = 'store sales',
       y = 'store area')
```

```r
store%>%
  ggplot(aes(Store_Sales,Items_Available ))+
  geom_point(fill='pink')+
  theme_bw()+
  labs(title='Items Available vs store area',
       x = 'Items Available',
       y = 'store area')
```

**Conclusion:**

Based on my observation, the store area and the number of different items available give a slight contribution to the number of sales for the store.

The significant correlation between the columns is the correlation between store area and the number of different items available. The wider the area of the store, the higher the number of items.
